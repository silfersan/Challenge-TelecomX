```python
import pandas as pd


dados = pd.read_json('/content/TelecomX_Data.json')


dados
```

```python
customer_df = pd.json_normalize(dados['customer'])
phone_df = pd.json_normalize(dados['phone'])
internet_df = pd.json_normalize(dados['internet'])
account_df = pd.json_normalize(dados['account'])
df = pd.concat([
    dados[['customerID', 'Churn']],
    customer_df,
    phone_df,
    internet_df,
    account_df
], axis=1)
print(dados.head())
```

```python
df = df.drop_duplicates()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('Não informado')
    else:
        df[col] = df[col].fillna(0)
```

```python
print(df['Churn'].dtype)
print(df['PaymentMethod'].dtype)
```

```python

df['Churn'] = df['Churn'].fillna('').str.strip().str.capitalize()
df['PaymentMethod'] = df['PaymentMethod'].fillna('').str.strip().str.lower()
```

```python
print("Valores ausentes em cada coluna:\n")
print(df.isnull().sum())
print("\nLinhas duplicadas:", df.duplicated().sum())
print("\nValores únicos das colunas categóricas:\n")
for col in df.select_dtypes(include='object'):
    print(f"Coluna {col}:")
    print(df[col].value_counts())
    print("\n")
```

```python
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
print()
```

```python
df['Churn'] = df['Churn'].astype(str).astype('category')
```

```python
import matplotlib.pyplot as plt

df['churn'] = df['churn'].replace({
    'No': 'Permaneceu',
    'Yes': 'Cancelou'
})

churn_counts = df['churn'].dropna().value_counts()
labels = churn_counts.index.tolist()
cores = ['#66c2a5', '#fc8d62'][:len(churn_counts)]
explode = [0.1 if i == churn_counts.max() else 0 for i in churn_counts]
def format_label(pct, allvals):
    absolute = int(round(pct / 100. * sum(allvals)))
    return f'{pct:.1f}%\n({absolute:,})'
plt.figure(figsize=(6, 6))
plt.pie(
    churn_counts,
    labels=labels,
    autopct=lambda pct: format_label(pct, churn_counts),
    startangle=90,
    colors=cores,
    explode=explode,
    shadow=True
)
plt.title('📉 Distribuição da Evasão de Clientes (Churn)', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def churn_por_categoria(coluna):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df, x=coluna, hue='churn', palette=['#66c2a5', '#fc8d62'])

    total = len(df)
    for p in ax.patches:
        altura = p.get_height()
        porcentagem = f'{100 * altura / total:.1f}%'
        ax.annotate(porcentagem,
                    (p.get_x() + p.get_width() / 2., altura),
                    ha='center', va='bottom', fontsize=9)

    plt.title(f'Distribuição de Evasão por {coluna.capitalize()}', fontsize=14)
    plt.xlabel(coluna.capitalize())
    plt.ylabel('Contagem de Clientes')
    plt.xticks(rotation=30)
    plt.legend(title='Situação')
    plt.tight_layout()
    plt.show()
churn_por_categoria('gender')
churn_por_categoria('contract')
churn_por_categoria('paymentmethod')
churn_por_categoria('internetservice')
churn_por_categoria('partner')
churn_por_categoria('multiplelines')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def churn_por_variavel_numerica(coluna):
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='churn', y=coluna, palette=['#66c2a5', '#fc8d62'])
    plt.title(f'{coluna.capitalize()} x Evasão (Churn)', fontsize=14)
    plt.xlabel('Churn')
    plt.ylabel(coluna.capitalize())
    plt.tight_layout()
    plt.show()
churn_por_variavel_numerica('tenure')
churn_por_variavel_numerica('charges.monthly')
churn_por_variavel_numerica('charges.total')
```

```python
df_corr = df.copy()
df_corr['churn_numeric'] = df_corr['churn'].map({'Permaneceu': 0, 'Cancelou': 1})
colunas_numericas = [
    'contas_diarias',
    'tenure',
    'charges.monthly',
    'charges.total',
    'churn_numeric',

]
corr_matrix = df_corr[colunas_numericas].corr()
print(corr_matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
plt.figure(figsize=(8, 5))
sns.boxplot(x='churn', y='contas_diarias', data=df)
plt.title('Distribuição da Conta Diária por Churn')
plt.show()
servicos = ['onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies']
df_corr['qtd_servicos'] = df_corr[servicos].apply(lambda x: x == 'Yes').sum(axis=1)
print(df_corr[['qtd_servicos', 'churn_numeric']].corr())
plt.figure(figsize=(8, 5))
sns.boxplot(x='churn', y='qtd_servicos', data=df_corr)
plt.title('Quantidade de Serviços Contratados por Situação de Churn')
plt.show()
```
